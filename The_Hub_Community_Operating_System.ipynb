```python
import google.generativeai as genai

genai.configure(api_key="")

model = genai.GenerativeModel("models/gemini-3.5-flash")

response = model.generate_content("Say hello")

print(response.text)
```

```python
def application_form():

    print("=== THE HUB BENGALURU ===")

    name = input("Name: ")

    skills = input("Skills: ")

    interests = input("Interests: ")

    goals = input("Goals: ")

    return {
        "name": name,
        "skills": skills,
        "interests": interests,
        "goals": goals
    }
```

```python
community_members = [

{
"name":"Rahul",
"role":"AI Startup Founder",
"skills":"AI SaaS, Business Strategy",
"interests":"Startups"
},

{
"name":"Ananya",
"role":"Content Creator",
"skills":"Podcasting, YouTube",
"interests":"Media"
},

{
"name":"Priya",
"role":"Product Manager",
"skills":"Product Strategy",
"interests":"AI Products"
},

{
"name":"Vikram",
"role":"Investor",
"skills":"Fundraising",
"interests":"Startups"
}

]
```

```python
events = [

"AI Builders Meetup",

"Founder Networking Night",

"Startup Pitch Event",

"Creator Workshop",

"Investor Connect Session"

]
```

```python
def profile_agent(user):

    prompt = f"""
    Analyze this Hub member.

    Name:
    {user['name']}

    Skills:
    {user['skills']}

    Interests:
    {user['interests']}

    Goals:
    {user['goals']}

    Generate profile summary.
    """

    result = model.generate_content(prompt)

    return result.text
```

```python
def interest_agent(profile):

    prompt = f"""
    Analyze profile.

    {profile}

    Extract:

    1. Primary Interests
    2. Networking Needs
    3. Collaboration Areas
    """

    result = model.generate_content(prompt)

    return result.text
```

```python
def matching_agent(profile, interests):

    prompt = f"""
    Profile:

    {profile}

    Interests:

    {interests}

    Members:

    {community_members}

    Recommend top 3 matches.
    """

    result = model.generate_content(prompt)

    return result.text
```

```python
def event_agent(profile, interests):

    prompt = f"""
    Profile:

    {profile}

    Interests:

    {interests}

    Events:

    {events}

    Recommend top events.
    """

    result = model.generate_content(prompt)

    return result.text
```

```python
def collaboration_agent(
    profile,
    matches,
    events_output
):

    prompt = f"""
    Profile:

    {profile}

    Matches:

    {matches}

    Events:

    {events_output}

    Suggest:

    Startup Ideas

    Collaborations

    Business Opportunities

    30 Day Action Plan
    """

    result = model.generate_content(prompt)

    return result.text
```

```python
import pandas as pd

def memory_agent(
    user,
    profile,
    interests,
    matches,
    event_results,
    collab
):

    data = pd.DataFrame({

    "Name":[user["name"]],

    "Profile":[profile],

    "Interests":[interests],

    "Matches":[matches],

    "Events":[event_results],

    "Collaboration":[collab]

    })

    data.to_csv(
        "hub_memory.csv",
        mode="a",
        header=False,
        index=False
    )

    print("Memory Saved")
```

```python
def run_community_os():

    user = application_form()

    print("\nRunning Profile Agent...\n")

    profile = profile_agent(user)

    print(profile)

    print("\nRunning Interest Agent...\n")

    interests = interest_agent(profile)

    print(interests)

    print("\nRunning Matching Agent...\n")

    matches = matching_agent(
        profile,
        interests
    )

    print(matches)

    print("\nRunning Event Agent...\n")

    event_results = event_agent(
        profile,
        interests
    )

    print(event_results)

    print("\nRunning Collaboration Agent...\n")

    collab = collaboration_agent(
        profile,
        matches,
        event_results
    )

    print(collab)

    memory_agent(
        user,
        profile,
        interests,
        matches,
        event_results,
        collab
    )

    print("\nSYSTEM COMPLETE")
```

```python
run_community_os()
```
